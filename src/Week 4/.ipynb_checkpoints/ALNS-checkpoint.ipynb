```python
# This is the ALNS initial framework code

import random
import math
import copy

# -----------------------------
# Basic data structure (can be connected according to Problem / Node)
# -----------------------------

class Solution:
    def __init__(self):
        self.truck_route = []      # Truck Routes (Customer Sequences)
        self.drone_assignments = {}  # Customer -> Is it delivered by drone?
        self.cost = float('inf')   # Total cost (time + energy)

    def evaluate(self, problem):
        # TODO: Calculate and update self.cost
        pass

    def copy(self):
        return copy.deepcopy(self)

    def remove_customers(self, customer_list):
        # TODO: Remove the customer (either truck or drone) from the current solution
        pass

    def insert_customer(self, customer, position=None):
        # TODO: Insert customer to truck or drone route
        pass

# -----------------------------
# Initial solution construction
# -----------------------------

def construct_initial_solution(problem):
    sol = Solution()
    # TODO: Construct a preliminary solution based on greedy or nearest neighbor
    return sol

# -----------------------------
# Destruction Operator
# -----------------------------

def random_removal(solution, num_remove):
    removed = random.sample(solution.truck_route, min(num_remove, len(solution.truck_route)))
    solution.remove_customers(removed)
    return removed


# -----------------------------
# Repair Operator
# -----------------------------

def greedy_repair(solution, removed_customers, problem):
    for customer in removed_customers:
        # TODO: Find the best insertion position or allocation method
        solution.insert_customer(customer)
    return solution

# -----------------------------
# Operator selection mechanism (roulette wheel)
# -----------------------------

def roulette_selection(weights):
    total = sum(weights.values())
    r = random.uniform(0, total)
    acc = 0
    for op, w in weights.items():
        acc += w
        if acc >= r:
            return op

# -----------------------------
# Simulated Annealing: Acceptance Criteria
# -----------------------------

def accept(new_sol, old_sol, temperature):
    if new_sol.cost < old_sol.cost:
        return True
    delta = new_sol.cost - old_sol.cost
    return random.random() < math.exp(-delta / temperature)

# -----------------------------
# ALNS Main Process
# -----------------------------

def run_alns(problem, max_iter=1000, num_remove=3):
    # initialization
    current = construct_initial_solution(problem)
    current.evaluate(problem)
    best = current.copy()

    temperature = 100.0
    cooling_rate = 0.995

    # Operator Pool
    destroy_ops = {
        'random': random_removal,
        # Adding more operators...
    }

    repair_ops = {
        'greedy': greedy_repair,
        # Adding more repair operators...
    }

    weights_destroy = {k: 1.0 for k in destroy_ops}
    weights_repair = {k: 1.0 for k in repair_ops}

    for it in range(max_iter):
        # Operator selection
        d_key = roulette_selection(weights_destroy)
        r_key = roulette_selection(weights_repair)

        # Uncopy & Manipulate
        new_sol = current.copy()
        removed = destroy_ops[d_key](new_sol, num_remove)
        new_sol = repair_ops[r_key](new_sol, removed, problem)
        new_sol.evaluate(problem)

        # Accept judgment
        if accept(new_sol, current, temperature):
            current = new_sol
        if new_sol.cost < best.cost:
            best = new_sol
            # Operator Score Rewards (Example)
            weights_destroy[d_key] += 1
            weights_repair[r_key] += 1

        # Cooling
        temperature *= cooling_rate

        if it % 50 == 0:
            print(f"[Iter {it}] Best cost: {best.cost:.2f}")

    return best
```

```python
# Solution.evaluate() needs to combine the multiple components provided in the previous weeks' tasks, 
# taking into account each task assignment (truck/drone), travel time, energy consumption, handover time, wind effect, waiting penalty, etc. 
# in the solution, and output a total cost function value.


class Solution:
    def __init__(self, assignments):
        """
        assignments: list of tuples (task_id, 'drone' or 'vehicle', handover_point)
        例如: [(0, 'drone', 'P1'), (1, 'vehicle', None), ...]
        """
        self.assignments = assignments
        self.cost = None

    def evaluate(self,
                 distance_drone, distance_vehicle,
                 time_drone, time_vehicle,
                 energy_drone, energy_vehicle,
                 drone_range_with_wind,
                 handover_time_dict,  # {(task_id, handover_point): time_in_seconds}
                 waiting_time_dict,  # {(task_id, handover_point): waiting_time_seconds}
                 penalty_waiting_time=1.0,
                 penalty_range_violation=1000.0):
        total_time = 0
        total_energy = 0
        total_penalty = 0

        for assign in self.assignments:
            task_id, mode, handover = assign

            if mode == 'drone':
                # time + energy
                total_time += time_drone[task_id]
                total_energy += energy_drone[task_id]

                # Range Penalty
                if distance_drone[task_id] > drone_range_with_wind[task_id]:
                    total_penalty += penalty_range_violation

                # Handover time + waiting penalty
                if handover:
                    total_time += handover_time_dict.get((task_id, handover), 0)
                    total_penalty += penalty_waiting_time * waiting_time_dict.get((task_id, handover), 0)

            elif mode == 'vehicle':
                total_time += time_vehicle[task_id]
                total_energy += energy_vehicle[task_id]

        # weight
        alpha = 0.5
        beta = 0.5

        # final cost
        self.cost = alpha * total_time + beta * total_energy + total_penalty
        return self.cost
```

```python
# Integrate the ALNS main framework code with the Solution.evaluate() function. 
# The overall goal is to allow each solution to correctly calculate the cost function for the ALNS acceptance criteria and update the optimal solution.

import copy
import random

class Solution:
    def __init__(self, assignments):
        self.assignments = assignments  # [(task_id, 'drone' or 'vehicle', handover_point)]
        self.cost = None

    def evaluate(self,
                 distance_drone, distance_vehicle,
                 time_drone, time_vehicle,
                 energy_drone, energy_vehicle,
                 drone_range_with_wind,
                 handover_time_dict,
                 waiting_time_dict,
                 penalty_waiting_time=1.0,
                 penalty_range_violation=1000.0):
        total_time = 0
        total_energy = 0
        total_penalty = 0

        for assign in self.assignments:
            task_id, mode, handover = assign

            if mode == 'drone':
                total_time += time_drone[task_id]
                total_energy += energy_drone[task_id]

                if distance_drone[task_id] > drone_range_with_wind[task_id]:
                    total_penalty += penalty_range_violation

                if handover:
                    total_time += handover_time_dict.get((task_id, handover), 0)
                    total_penalty += penalty_waiting_time * waiting_time_dict.get((task_id, handover), 0)

            elif mode == 'vehicle':
                total_time += time_vehicle[task_id]
                total_energy += energy_vehicle[task_id]

        alpha = 0.5
        beta = 0.5
        self.cost = alpha * total_time + beta * total_energy + total_penalty
        return self.cost

# === ALNS Main Process ===
def ALNS(initial_solution, 
         destroy_ops, repair_ops, 
         iterations,
         distance_drone, distance_vehicle,
         time_drone, time_vehicle,
         energy_drone, energy_vehicle,
         drone_range_with_wind,
         handover_time_dict, waiting_time_dict):
    
    current_solution = copy.deepcopy(initial_solution)
    current_cost = current_solution.evaluate(
        distance_drone, distance_vehicle,
        time_drone, time_vehicle,
        energy_drone, energy_vehicle,
        drone_range_with_wind,
        handover_time_dict, waiting_time_dict
    )

    best_solution = copy.deepcopy(current_solution)
    best_cost = current_cost

    weights = {d: 1 for d in destroy_ops}
    weights.update({r: 1 for r in repair_ops})

    for i in range(iterations):
        # 1. choice
        destroy = random.choices(destroy_ops, weights=[weights[d] for d in destroy_ops])[0]
        repair = random.choices(repair_ops, weights=[weights[r] for r in repair_ops])[0]

        # 2. Application Destroy-Repair
        partial_solution = destroy(current_solution)
        new_solution = repair(partial_solution)

        # 3. Evaluation Cost
        new_cost = new_solution.evaluate(
            distance_drone, distance_vehicle,
            time_drone, time_vehicle,
            energy_drone, energy_vehicle,
            drone_range_with_wind,
            handover_time_dict, waiting_time_dict
        )

        # 4. Accept new interpretations
        if new_cost < current_cost:
            current_solution = new_solution
            current_cost = new_cost

            if new_cost < best_cost:
                best_solution = copy.deepcopy(new_solution)
                best_cost = new_cost

        print(f"[Iteration {i+1}] Current price: {current_cost:.2f} | Optimal Price: {best_cost:.2f}")

    return best_solution
```

```python
# Here are simple but representative destroy and repair operator functions, make sure they work with your Solution class and the ALNS main framework.

import random

# Randomly remove the assignment of some tasks (e.g. remove 20%)
def random_destroy(solution, destroy_fraction=0.2):
    num_remove = int(len(solution.assignments) * destroy_fraction)
    new_assignments = solution.assignments.copy()

    to_remove = random.sample(range(len(new_assignments)), num_remove)
    for idx in sorted(to_remove, reverse=True):
        del new_assignments[idx]

    return Solution(new_assignments)
```

```python
# Remove only the tasks assigned by the drone (simulating sudden changes in wind speed or limited battery life)
def remove_drones_only(solution, destroy_fraction=0.3):
    drone_tasks = [a for a in solution.assignments if a[1] == 'drone']
    num_remove = int(len(drone_tasks) * destroy_fraction)
    
    to_remove = random.sample(drone_tasks, num_remove)
    new_assignments = [a for a in solution.assignments if a not in to_remove]
    
    return Solution(new_assignments)
```

```python
# Randomly reassign unassigned tasks (randomly assign drone or vehicle)
def random_repair(partial_solution, all_task_ids, handover_points):
    assigned_tasks = {a[0] for a in partial_solution.assignments}
    unassigned_tasks = [t for t in all_task_ids if t not in assigned_tasks]
    new_assignments = partial_solution.assignments.copy()

    for task_id in unassigned_tasks:
        mode = random.choice(['drone', 'vehicle'])
        if mode == 'drone':
            handover = random.choice(handover_points)
        else:
            handover = None
        new_assignments.append((task_id, mode, handover))

    return Solution(new_assignments)
```

```python
# Try to prioritize unfinished tasks with drones (a more "aggressive" strategy)
def greedy_repair_drone_first(partial_solution, all_task_ids, handover_points):
    assigned_tasks = {a[0] for a in partial_solution.assignments}
    unassigned_tasks = [t for t in all_task_ids if t not in assigned_tasks]
    new_assignments = partial_solution.assignments.copy()

    for task_id in unassigned_tasks:
        # Try the drone first (lower energy consumption), otherwise fallback to the car
        if random.random() < 0.7: # 70% chance to use drone
            handover = random.choice(handover_points)
            new_assignments.append((task_id, 'drone', handover))
        else:
            new_assignments.append((task_id, 'vehicle', None))

    return Solution(new_assignments)
```

```python
# Integrating all the above functions, the minimum runnable version of the ALNS framework sample includes:
# Virtual task points T0 to T19
# Virtual distance/time/energy matrix
# Solution structure
# 2 destroy operators + 2 repair operators
# ALNS main loop (20 rounds)

import random
import math
import matplotlib.pyplot as plt

# ========== Step 1: Generate fake data ==========

random.seed(42)
task_ids = list(range(20))
handover_points = ['P1', 'P2', 'P3']

# Simulate distance, time, and energy matrices (all units are pseudo-random numbers)
distance_dict = {f'T{i}': {f'T{j}': random.uniform(1, 10) for j in task_ids} for i in task_ids}
time_dict = {f'T{i}': {f'T{j}': distance_dict[f'T{i}'][f'T{j}'] / 10 + random.uniform(0.5, 1.0) for j in task_ids} for i in task_ids}
energy_dict = {f'T{i}': {f'T{j}': distance_dict[f'T{i}'][f'T{j}'] * 0.5 + random.uniform(0.1, 1.0) for j in task_ids} for i in task_ids}

# Store the time and energy matrices as lists to avoid calculation each time
def get_time_energy(task_from, task_to):
    time = time_dict[task_from][task_to]
    energy = energy_dict[task_from][task_to]
    return time, energy

# ========== Step 2: Define Solution class and cost function ==========

class Solution:
    def __init__(self, assignments):
        self.assignments = assignments
        # Pre-calculate the time and energy consumption between tasks to avoid repeated calculations
        self.precomputed_costs = self.compute_costs(assignments)

    def compute_costs(self, assignments):
        costs = {'truck_time': 0, 'truck_energy': 0, 'drone_time': 0, 'drone_energy': 0, 'penalty': 0}
        for assignment in assignments:
            task_id, mode, handover_point = assignment
            from_id = f'T{task_id}'
            to_id = f'T{(task_id + 1) % len(task_ids)}'

            time, energy = get_time_energy(from_id, to_id)

            if mode == 'vehicle':
                costs['truck_time'] += time
                costs['truck_energy'] += energy
            elif mode == 'drone':
                if handover_point is None:
                    costs['penalty'] += 1000
                costs['drone_time'] += time
                costs['drone_energy'] += energy
        
        return costs

    def evaluate(self):
        total_cost = sum(self.precomputed_costs.values())
        return total_cost

# ========== Step 3: Constructing an initial solution ==========

def generate_initial_solution():
    assignments = []
    for task_id in task_ids:
        mode = random.choice(['drone', 'vehicle'])
        if mode == 'drone':
            handover = random.choice(handover_points)
        else:
            handover = None
        assignments.append((task_id, mode, handover))
    return Solution(assignments)

# ========== Step 4: Destroy & Repair Operations ==========

def random_destroy(solution, destroy_fraction=0.3):
    num_remove = int(len(solution.assignments) * destroy_fraction)
    new_assignments = solution.assignments.copy()
    to_remove = random.sample(range(len(new_assignments)), num_remove)
    for idx in sorted(to_remove, reverse=True):
        del new_assignments[idx]
    return Solution(new_assignments)

def remove_drones_only(solution, destroy_fraction=0.5):
    drone_tasks = [a for a in solution.assignments if a[1] == 'drone']
    num_remove = int(len(drone_tasks) * destroy_fraction)
    to_remove = random.sample(drone_tasks, min(num_remove, len(drone_tasks)))
    new_assignments = [a for a in solution.assignments if a not in to_remove]
    return Solution(new_assignments)

def random_repair(partial_solution, all_task_ids, handover_points):
    assigned = {a[0] for a in partial_solution.assignments}
    unassigned = [t for t in all_task_ids if t not in assigned]
    new_assignments = partial_solution.assignments.copy()

    for t in unassigned:
        mode = random.choice(['drone', 'vehicle'])
        h = random.choice(handover_points) if mode == 'drone' else None
        new_assignments.append((t, mode, h))
    return Solution(new_assignments)

def greedy_repair_drone_first(partial_solution, all_task_ids, handover_points):
    assigned = {a[0] for a in partial_solution.assignments}
    unassigned = [t for t in all_task_ids if t not in assigned]
    new_assignments = partial_solution.assignments.copy()

    for t in unassigned:
        if random.random() < 0.7:
            h = random.choice(handover_points)
            new_assignments.append((t, 'drone', h))
        else:
            new_assignments.append((t, 'vehicle', None))
    return Solution(new_assignments)

# Mode of exchanging tasks
def swap_tasks(solution):
    new_assignments = solution.assignments.copy()
    idx1, idx2 = random.sample(range(len(new_assignments)), 2)
    new_assignments[idx1] = (new_assignments[idx1][0], 'vehicle' if new_assignments[idx1][1] == 'drone' else 'drone', new_assignments[idx1][2])
    new_assignments[idx2] = (new_assignments[idx2][0], 'vehicle' if new_assignments[idx2][1] == 'drone' else 'drone', new_assignments[idx2][2])
    return Solution(new_assignments)

# Deleting multiple tasks
def multi_task_destroy(solution, destroy_fraction=0.3):
    num_remove = int(len(solution.assignments) * destroy_fraction)
    to_remove = random.sample(range(len(solution.assignments)), num_remove)
    new_assignments = [assignment for idx, assignment in enumerate(solution.assignments) if idx not in to_remove]
    return Solution(new_assignments)

# Reorder tasks
def reorder_tasks(solution):
    new_assignments = solution.assignments.copy()
    random.shuffle(new_assignments)
    return Solution(new_assignments)

# ========== Step 5: ALNS Main Flow ==========

def ALNS(iterations=20):
    current = generate_initial_solution()
    best = current
    best_cost = best.evaluate()

    # Added new destruction operations
    destroy_ops = [
        lambda s: random_destroy(s, 0.3), 
        lambda s: remove_drones_only(s, 0.4),
        lambda s: swap_tasks(s),  # Added exchange mission mode
        lambda s: multi_task_destroy(s, 0.3),  # Add and delete multiple tasks
        lambda s: reorder_tasks(s),  # Added reordering tasks
    ]

    repair_ops = [lambda s: random_repair(s, task_ids, handover_points),
                  lambda s: greedy_repair_drone_first(s, task_ids, handover_points)]

    # Prepare for plotting
    costs = []

    for i in range(iterations):
        destroy = random.choice(destroy_ops)
        repair = random.choice(repair_ops)

        destroyed = destroy(current)
        repaired = repair(destroyed)
        cost = repaired.evaluate()

        # Store cost for plotting
        costs.append(cost)

        print(f"[{i+1:02d}] Cost: {cost:.2f} | Best: {best_cost:.2f} | Temperature: {100 - (i * 5)}")

        if cost < best_cost:
            best = repaired
            best_cost = cost
            current = repaired
        else:
            # Probability of accepting a worse solution (with simulated annealing)
            if random.random() < 0.1:
                current = repaired

    # Plot cost over iterations
    plt.plot(range(1, iterations + 1), costs, marker='o', color='b', label="Cost")
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('ALNS Cost Evolution')
    plt.grid(True)
    plt.legend()
    plt.show()

    print("\n✅ Optimal Cost：", best_cost)

if __name__ == "__main__":
    ALNS()
```
